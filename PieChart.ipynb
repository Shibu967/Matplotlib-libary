```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
slices=[70,30]
plt.pie(slices)
plt.title("First pie Chart")
```

# Need labels for

```python
slices=[70,30]
labels=['Attempted','left']
plt.pie(slices,labels=labels)
plt.title("First pie Chart")
```

# lets change the colors of pie

```python
slices=[70,30,20,40]
labels=['correct','Wrong','Reviewed','Left']
colors=['red','blue','green','yellow']
plt.pie(slices,labels=labels,colors=colors)
plt.title("Firs Pie Chart")
```

# Adding colors with RBG hex values

```python
slices=[70,30,20,40]
labels=['correct','Wrong','Reviewed','Left']
colors=['#fc4f30','#008fd5','#6d904f','#e5ae37']
plt.pie(slices,labels=labels,colors=colors)
plt.title("First Pie Chart")
```

# Want to give emphasis on item

```python
slices=[70,30,20,40]
labels=['correct','Wrong','Reviewed','Left']
colors=['#fc4f30','#008fd5','#6d904f','#e5ae37']
explode=[0,0.1,0,0]
plt.pie(slices,labels=labels,colors=colors,explode=explode)
plt.title("First Pie Chart")
```

```python
slices=[70,30,20,40]
labels=['correct','Wrong','Reviewed','Left']
colors=['#fc4f30','#008fd5','#6d904f','#e5ae37']
explode=[0,0,0,0.1]
plt.pie(slices,labels=labels,colors=colors,explode=explode,shadow=True)
plt.title("First Pie Chart")
```

```python
slices=[70,30,20,40]
labels=['correct','Wrong','Reviewed','Left']
colors=['#fc4f30','#008fd5','#6d904f','#e5ae37']
explode=[0,0,0,0.1]
plt.pie(slices,labels=labels,colors=colors,explode=explode,shadow=True,autopct='%1.1f%%')
plt.title("First Pie Chart")
```
